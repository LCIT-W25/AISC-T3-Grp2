```python
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
```

```python
# Ensure required NLTK resources are downloaded
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
# Load the dataset
file_path = 'Data NLP.csv'
data = pd.read_csv(file_path)
```

```python
# Column containing text
data_column = 'Data'

# Display the first few rows of the data
print("First few rows of the dataset:")
print(data.head())

# Display the entire dataset if needed (caution with large datasets)
print("\nComplete dataset:")
print(data)
```

```python
# 1. Records with a date without alphabets
pattern_date_no_alpha = r'^\d{2}[/-]\d{2}[/-]\d{4}$'
date_no_alpha_count = data[data[data_column].str.contains(pattern_date_no_alpha, na=False)].shape[0]
```

```python
print(f"Records with a date without alphabets: {date_no_alpha_count}")
```

```python
# 2. Records with words starting with 'w'
pattern_word_w = r'\b[wW]\w*\b'
word_w_count = data[data[data_column].str.contains(pattern_word_w, na=False)].shape[0]
```

```python
print(f"Records with words starting with 'w': {word_w_count}")
```

```python
# 3. Records with a word starting with an alphabet and not a URL
pattern_word_alpha_no_url = r'\b[a-zA-Z](?![a-zA-Z]*://)\w*\b'
word_alpha_no_url_count = data[data[data_column].str.contains(pattern_word_alpha_no_url, na=False)].shape[0]
```

```python
print(f"Records with a word starting with an alphabet and not a URL: {word_alpha_no_url_count}")
```

```python
# 4. Tweets with specified emojis
pattern_emojis = r':\)|:D|;\)|:P'
emoji_count = data[data[data_column].str.contains(pattern_emojis, na=False)].shape[0]
```

```python
print(f"Tweets with specified emojis: {emoji_count}")
```

5. Records containing decimal numbers

```python
pattern_decimal = r'\b\d+\.\d+\b'
decimal_count = data[data[data_column].str.contains(pattern_decimal, na=False)].shape[0]
```

```python
print(f"Records containing decimal numbers: {decimal_count}")
```

6. Total number of IP addresses

```python
pattern_ip = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'
ip_addresses = data[data_column].str.findall(pattern_ip).dropna()
total_ip_count = sum(ip_addresses.map(len))
```

```python
print(f"Total number of IP addresses: {total_ip_count}")
```

```python
print("Extracted IP addresses:")
print(ip_addresses)
```

```python
# 7. Records with a new line
pattern_new_line = r'\n'
new_line_count = data[data[data_column].str.contains(pattern_new_line, na=False)].shape[0]
```

```python
print(f"Records with a new line: {new_line_count}")
```

```python
# 8. Total number of hashtags
pattern_hashtags = r'#\w+'
hashtags = data[data_column].str.findall(pattern_hashtags).dropna()
total_hashtags_count = sum(hashtags.map(len))
```

```python
print(f"Total number of hashtags: {total_hashtags_count}")
```

```python
# 9. Substitute all non-alphanumeric characters with a new line
substituted_data = data[data_column].str.replace(r'[^a-zA-Z0-9]', '\n', regex=True)
```

```python
# Print the substituted data
print("Substituted data (non-alphanumeric characters replaced with new line):")
print(substituted_data)
```

```python
# 10. Total number of URLs
pattern_url = r'http[s]?://\S+'
urls = data[data_column].str.findall(pattern_url).dropna()
total_urls_count = sum(urls.map(len))
```

```python
print(f"Total number of URLs: {total_urls_count}")
```

# B. Stemming and Lemmatization

```python
# Tokenization
all_tokens = data[data_column].str.split().explode().dropna()
unique_tokens_before = set(all_tokens)
```

```python
# Print the unique tokens before any further processing
print(f"Unique tokens before processing: {unique_tokens_before}")
```

```python
# Print the number of unique tokens before processing
print(f"Number of unique tokens before processing: {len(unique_tokens_before)}")
```

```python
# 1. Stemming
stemmer = PorterStemmer()
stemmed_tokens = all_tokens.map(stemmer.stem)
unique_tokens_after_stemming = set(stemmed_tokens)
```

```python
print(f"unique_tokens_after_stemming: {len(unique_tokens_after_stemming)}")
```

```python
# 2. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = all_tokens.map(lemmatizer.lemmatize)
unique_tokens_after_lemmatization = set(lemmatized_tokens)
```

```python
print(f"unique_tokens_after_lemmatization: {len(unique_tokens_after_lemmatization)}")
```

```python
# 3. Word frequency comparison
stemmed_frequency = Counter(stemmed_tokens)
lemmatized_frequency = Counter(lemmatized_tokens)
```

```python
print(f"top_10_stemmed_words: {stemmed_frequency.most_common(10)}")
print(f"top_10_lemmatized_words: {lemmatized_frequency.most_common(10)}")
```

```python
# 4. Change in word frequencies after stop word removal and normalization
stop_words = set(stopwords.words('english'))
tokens_without_stopwords = [token for token in all_tokens if token.lower() not in stop_words]
normalized_tokens = [token.lower() for token in tokens_without_stopwords]
normalized_frequency = Counter(normalized_tokens)
```

```python
print(f"top_10_lemmatized_words: {lemmatized_frequency.most_common(10)}")
print(f"normalized_frequency_top_10: {normalized_frequency.most_common(10)}")
```
