```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('Data.csv')
```

```python
#nltk.download('corpus')
#nltk.download('wordnet')
#nltk.download('stopwords')
```

```python
stemmer = PorterStemmer()
def stem_text(text):
    tokens = word_tokenize(text)
    return ' '.join([stemmer.stem(token) for token in tokens])
```

```python
df['stemmed'] = df['Data'].apply(stem_text)
```

```python
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    tokens = word_tokenize(text)
    return ' '.join(lemmatizer.lemmatize(token) for token in tokens)
```

```python
df['lemmatized'] = df['Data'].apply(lemmatize_text)
```

```python
def unique_word_count(text):
    tokens = word_tokenize(text)
    return len(set(tokens))
```

```python
unique_word_count(' '.join(df['Data']))
```

```python
unique_word_count(' '.join(df['stemmed']))
```

```python
unique_word_count(' '.join(df['lemmatized']))
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    tokens = word_tokenize(text)
    return ' '.join([word for word in tokens if word not in stop_words])
```

```python
df['no_stopwords'] = df['Data'].apply(remove_stopwords)
```

```python
Counter((' '.join(df['no_stopwords'])).split()).most_common(10)
```
