## Disjunctions
- **Example**: `[wW]oodchuck`  
  Matches any word with `Woodchuck` or `woodchuck`.

## Ranges
- `[A-Z]`: An uppercase letter.  
- `[a-z]`: A lowercase letter.  
- `[0-9]`: A single digit.

## Negation in Disjunction
- `[^Ss]`: **Carat (`^`)** means negation only when first inside `[]`.  
- `[^A-Z]`: Not an uppercase letter.  
- `[^e^]`: Not `e` nor `^`.  
  - Finds the first character that satisfies this condition.

## Pipes `|` for Disjunction
- `groundhog|woodchuck`: Matches either "groundhog" or "woodchuck".  
- `yours|mine`: Matches either "yours" or "mine".

## Characters
- `?`: The previous character is optional.  
- `*`: Zero or more of the previous character.  
- `+`: One or more of the previous character (must appear at least once).  
- `.`: A wildcard — can match any character.

## Special Anchors
- `^`: Carat at the start — matches if the string starts with the condition.  
- `$`: Matches the **end** of the string.  
- `\.`: The backslash (`\`) allows searching for a special character as a literal.  
  - Example: `\.$` matches a literal period (`.`) at the end of the string.

## 1. How many records have a date that is expressed without using 
alphabets?

```python
import re

# Assuming "dates without alphabets" are sequences like '01/01/2023' or '2023-12-31'
date_no_alphabets_pattern = r'\b\d{1,2}[-/\\]\d{1,2}[-/\\]\d{2,4}\b'
records_with_dates_no_alphabets = data['Data'].str.contains(date_no_alphabets_pattern, regex=True, na=False)
# Count records matching the date pattern
date_count = records_with_dates_no_alphabets.sum()
```

## B. Perform stemming and lemmatization (40%)

## 1. Use porter stemmer to run stemming. Count the number of unique words/tokens before and after stemming

Stemming is the process of producing morphological variants of a root/base word.

```python
#!pip install nltk
import nltk
#nltk.download('punkt_tab')
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()

# Tokenize the words
data['Tokens'] = data['Data'].apply(lambda x: word_tokenize(str(x)))
all_tokens = [token for tokens in data['Tokens'] for token in tokens]
unique_tokens_before = set(all_tokens)
```

### Before stemming

```python
len(unique_tokens_before)
```

```python
list(unique_tokens_before)[:5]
```

```python
#Perform Stemming
stemmed_tokens = [ps.stem(token) for token in all_tokens]
unique_tokens_after_stemming = set(stemmed_tokens)
```

### After stemming

```python
len(unique_tokens_after_stemming)
```

```python
list(unique_tokens_after_stemming)[:5]
```

## 2. Perform lemmatization using NLTK lemmatizer. Count the number of unique words/tokens before and after lemmatization

 ### lemmatization seeks to distill words to their foundational forms. In this linguistic refinement, the resultant base word is referred to as a “lemma.”

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in all_tokens]
```

```python
unique_tokens_after_lemmatization = set(lemmatized_tokens)
```

```python
len(unique_tokens_after_lemmatization)
```

```python
list(unique_tokens_after_lemmatization)[:5]
```

## 3. Compare the change in word frequencies from stemming andlemmatization. Which are the top 10 words after stemming/lemmatization

```python
stemmed_dict = {}
lemmatized_dict = {}
for token in stemmed_tokens:
    stemmed_dict[token] = stemmed_dict.get(token, 0) + 1
for token in lemmatized_tokens:
    lemmatized_dict[token] = lemmatized_dict.get(token, 0) + 1
```

```python
top_stemmed = sorted(stemmed_dict.items(), key=lambda x: x[1], reverse=True)[:10]
top_stemmed
```

```python
top_lemmatized = sorted(lemmatized_dict.items(), key=lambda x: x[1], reverse=True)[:10]
top_lemmatized
```

## 4. What is the change in word frequencies if normalization is done after stop word removal.

```python
from nltk.corpus import stopwords
#nltk.download("stopwords")
```

```python
stop_word_removal = [token for token in unique_tokens_before if token not in stopwords.words('english')]
len(stop_word_removal)
```

```python
stemmed_tokens_wostopw = [ps.stem(token) for token in stop_word_removal]
lemmatized_tokens_wostopw = [lemmatizer.lemmatize(token) for token in stop_word_removal]
```

```python
stemmed_dict_wostopw = {}
lemmatized_dict_wostopw = {}
for token in stemmed_tokens_wostopw:
    stemmed_dict_wostopw[token] = stemmed_dict_wostopw.get(token, 0) + 1
for token in lemmatized_tokens_wostopw:
    lemmatized_dict_wostopw[token] = lemmatized_dict_wostopw.get(token, 0) + 1
```

```python
top_stemmed_wostopw = sorted(stemmed_dict_wostopw.items(), key=lambda x: x[1], reverse=True)[:10]
top_stemmed_wostopw 
```

```python
top_lemmatized_wostopw = sorted(lemmatized_dict_wostopw.items(), key=lambda x: x[1], reverse=True)[:10]
top_lemmatized_wostopw
```

If we remove the stopwords before normalizing with stemming or lemmatization, we can see that the resulting dictionaries and top frequency words change, in the case of lemmatization we noticed that the most that a word appears on the twitts is twice.
