# PART A

```python
# Use the 'Data' column for text processing
text_column = df['Data']

# Define a function to count matches for regex patterns
def count_matches(pattern, text_column):
    return sum(bool(re.search(pattern, text)) for text in text_column)
```

```python
# 1. Records with a date expressed without using alphabets
date_without_alphabets = r'\b\d{1,4}[-/]\d{1,2}[-/]\d{2,4}\b'
count_dates_no_alphabets = count_matches(date_without_alphabets, text_column)
print(f"1. Records with dates without alphabets: {count_dates_no_alphabets}")
```

```python
# 2. Records with a word starting with the letter "w"
word_starting_with_w = r'\bw\w*\b'
count_words_with_w = count_matches(word_starting_with_w, text_column)
print(f"2. Records with words starting with 'w': {count_words_with_w}")
```

```python
word_not_url = r'\b(?!http|www)[a-zA-Z][a-zA-Z0-9]*\b'
count_words_not_url = count_matches(word_not_url, text_column)
print(f"3. Words starting with an alphabet and not URLs: {count_words_not_url}")
```

```python
emojis = r':\)|:D|;\)|:P'
count_emojis = count_matches(emojis, text_column)
print(f"4. Tweets with specific emojis: {count_emojis}")
```

```python
# 5. Records containing a decimal number
decimal_number = r'\b\d+\.\d+\b'
count_decimals = count_matches(decimal_number, text_column)
print(f"5. Records with decimal numbers: {count_decimals}")
```

```python
# 6. Total number of IP addresses
ip_address = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'
total_ips = sum(len(re.findall(ip_address, text)) for text in text_column)
print(f"6. Total IP addresses: {total_ips}")
```

```python
# 7. Records containing a new line
new_line = r'\n'
count_new_lines = count_matches(new_line, text_column)
print(f"7. Records with new lines: {count_new_lines}")
```

```python
# 8. Total number of hashtags across all tweets
hashtag = r'#\w+'
total_hashtags = sum(len(re.findall(hashtag, text)) for text in text_column)
print(f"8. Total hashtags: {total_hashtags}")
```

```python
# 9. Substitute all non-alphanumeric characters with a new line
df['Text_with_new_lines'] = text_column.apply(lambda x: re.sub(r'[^a-zA-Z0-9]', '\n', x))
print("9. Code to substitute non-alphanumeric characters with a new line is executed.")
```

```python
# 10. Total number of URLs across all tweets
urls = r'\bhttps?://[^\s]+\b'
total_urls = sum(len(re.findall(urls, text)) for text in text_column)
print(f"10. Total URLs: {total_urls}")
```

```python
# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
# Tokenizing text and cleaning
cleaned_text = text_column.apply(lambda tokens: [word.lower() for word in tokens if word.isalnum()])

# Applying Porter Stemmer
porter_stemmer = PorterStemmer()
stemmed_text = cleaned_text.apply(lambda tokens: [porter_stemmer.stem(word) for word in tokens])

# Counting unique words before and after stemming
unique_words_before_stemming = set([word for tokens in cleaned_text for word in tokens])
unique_words_after_stemming = set([word for tokens in stemmed_text for word in tokens])

print(f"Unique words before stemming: {len(unique_words_before_stemming)}")
print(f"Unique words after stemming: {len(unique_words_after_stemming)}")
```

```python
# Applying Lemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_text = cleaned_text.apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])

# Counting unique words before and after lemmatization
unique_words_before_lemmatization = set([word for tokens in cleaned_text for word in tokens])
unique_words_after_lemmatization = set([word for tokens in lemmatized_text for word in tokens])

print(f"Unique words before lemmatization: {len(unique_words_before_lemmatization)}")
print(f"Unique words after lemmatization: {len(unique_words_after_lemmatization)}")
```

```python
from collections import Counter

# Counting word frequencies after stemming
stemmed_words = [word for tokens in stemmed_text for word in tokens]
stemmed_word_freq = Counter(stemmed_words)

# Counting word frequencies after lemmatization
lemmatized_words = [word for tokens in lemmatized_text for word in tokens]
lemmatized_word_freq = Counter(lemmatized_words)

# Getting the top 10 most frequent words
top_10_stemmed = stemmed_word_freq.most_common(10)
top_10_lemmatized = lemmatized_word_freq.most_common(10)

print(f"Top 10 words after stemming: {top_10_stemmed}")
print(f"Top 10 words after lemmatization: {top_10_lemmatized}")
```

```python
# Remove stopwords and normalize (convert to lowercase and keep alphanumeric words only)
stop_words = set(stopwords.words('english'))
normalized_text = cleaned_text.apply(lambda tokens: [word for word in tokens if word not in stop_words])

# Counting word frequencies after stopword removal and normalization
normalized_words = [word for tokens in normalized_text for word in tokens]
normalized_word_freq = Counter(normalized_words)

# Display the top 10 most common words after normalization
print(f"Top 10 words after stopword removal and normalization: {normalized_word_freq.most_common(10)}")
```
